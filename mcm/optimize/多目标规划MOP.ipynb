# 多目标规划 MOP

## 模型

考虑相互影响 (常常导致无法同时最优) 的多目标的规划问题

### 解の分类

- 最优解: 全最优 (很有可能不存在)

- 有效解: 至少一个最优，非全最优

- 满意解: 全部目标在给定阈值内

### 解决方法

多目标 $\rightarrow$ 单目标

- 线性加权法

    线性加权得到单目标

    注意要先无量纲化/归一化

- $\epsilon$ 约束法

    1. 根绝决策者偏好，选择一个主要目标作单目标

    2. 其他目标给定 $\epsilon$ 作阈值放入约束条件

- 理想点法

    取到每个**单目标最优点**の线性加权平方和为单目标

- 优先级法

    1. 界定多目标优先级

    2. 先保证高优先级目标最优

    3. 在此前提下再取低优先级目标最优

### 敏感性分析

检验**线性加权法**中权重的调整影响

画 **"$f_{n}$ 权重 - $x_{1} / x_{2}$"** 图像观察突变点
